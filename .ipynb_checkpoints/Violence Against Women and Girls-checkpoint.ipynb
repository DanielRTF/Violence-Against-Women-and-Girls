### Now we will create 2 new columns, where in the first one we will clasify the countries into regions, and the second one the majority religion in that country. 

```python
regions = {
    'South America': ['Bolivia', 'Colombia', 'Guyana', 'Peru'],
    'Middle East': ['Armenia', 'Azerbaijan', 'Egypt', 'Jordan', 'Turkey', 'Yemen'],
    'Africa': ['Angola', 'Benin', 'Burkina Faso', 'Burundi', 'Cameroon', 'Chad', 'Comoros', 'Congo', 
               'Congo Democratic Republic', "Cote d'Ivoire", 'Eritrea', 'Eswatini', 'Ethiopia', 'Gabon', 
               'Gambia', 'Ghana', 'Guinea', 'Kenya', 'Lesotho', 'Liberia', 'Madagascar', 'Malawi', 'Mali', 'Morocco', 
               'Mozambique', 'Namibia', 'Niger', 'Nigeria', 'Rwanda', 'Sao Tome and Principe', 'Senegal', 
               'Sierra Leone', 'South Africa', 'Tanzania', 'Togo', 'Uganda', 'Zambia', 'Zimbabwe'],
    'Asia': ['Afghanistan', 'Bangladesh', 'Cambodia', 'India', 'Indonesia', 'Kyrgyz Republic', 'Maldives', 
             'Myanmar', 'Nepal', 'Pakistan', 'Philippines', 'Tajikistan', 'Timor-Leste', 'Turkmenistan'],
    'Balkans': ['Albania', 'Moldova'],
    'Other': ['Dominican Republic', 'Guatemala', 'Haiti', 'Honduras', 'Ukraine', 'Nicaragua']
}

majority_religions = {
    'Muslim': ['Afghanistan', 'Albania', 'Azerbaijan', 'Bangladesh', 'Chad', 'Comoros', 'Egypt', 'Gambia', 
               'Indonesia', 'Jordan', 'Kyrgyz Republic', 'Maldives', 'Mali', 'Morocco', 'Niger', 'Pakistan', 
               'Senegal', 'Sierra Leone', 'Tajikistan', 'Turkey', 'Turkmenistan', 'Yemen'],
    'Christian': ['Angola', 'Armenia', 'Benin', 'Bolivia', 'Burkina Faso', 'Burundi', 'Cameroon', 'Colombia', 
                  'Congo', 'Congo Democratic Republic', "Cote d'Ivoire", 'Dominican Republic', 'Eritrea', 
                  'Eswatini', 'Ethiopia', 'Gabon', 'Ghana', 'Guatemala', 'Guinea', 'Guyana', 'Haiti', 'Honduras', 
                  'Kenya', 'Lesotho', 'Liberia', 'Madagascar', 'Malawi', 'Moldova', 'Mozambique', 'Namibia', 'Nicaragua', 'Nigeria', 
                  'Peru', 'Philippines', 'Rwanda', 'Sao Tome and Principe', 'South Africa', 'Tanzania', 'Timor-Leste', 
                  'Togo', 'Uganda', 'Ukraine', 'Zambia', 'Zimbabwe'],
    'Other': ['Cambodia', 'India', 'Myanmar', 'Nepal']
}

def get_region(country):
    for region, countries in regions.items():
        if country in countries:
            return region
    return 'Unknown'

def get_religion(country):
    for religion, countries in majority_religions.items():
        if country in countries:
            return religion
    return 'Unknown'

female_residence['Region'] = female_residence['Country'].apply(get_region)
female_residence['Majority Religion'] = female_residence['Country'].apply(get_religion)

display(female_residence.head())
```

```python
display(female_residence['Majority Religion'].unique())
display(female_residence['Region'].unique())
```

# Exploratory Data Analysis <a class="anchor" id="EDA"></a>

```python
display(female_residence['Question'].unique())
```

```python
question_ids = {
    '... if she burns the food': 1,
    '... for at least one specific reason': 2,
    '... if she argues with him': 3,
    '... if she goes out without telling him': 4,
    '... if she neglects the children': 5,
    '... if she refuses to have sex with him': 6
}

female_residence['Question_ID'] = female_residence['Question'].map(question_ids)
female_residence.head(20)
```

```python
# Group by Region, Residence, and Question_ID to calculate the mean Value
grouped = female_residence.groupby(['Region', 'Demographics Response', 'Question_ID'])['Value'].mean().reset_index()
```

```python
# Pivot the table to have Question_ID as columns
pivot_table = grouped.pivot_table(index=['Region', 'Demographics Response'], columns='Question_ID', values='Value')
```

```python
# Plot the data
pivot_table.plot(kind='bar', figsize=(12, 8))
plt.title('Average Percentage of Women Agreeing with Each Question by Region and Residence')
plt.xlabel('Region and Residence')
plt.ylabel('Average Percentage')
plt.legend(title='Question ID')
plt.show()
```

# Visualization <a class="anchor" id="Vis"></a>
This graph is not easy to read, so we will a graph per region. 

```python
# Loop through each unique region to create separate plots
unique_regions = female_residence['Region'].unique()

for region in unique_regions:
    region_data = grouped[grouped['Region'] == region]
    pivot_table = region_data.pivot_table(index='Demographics Response', columns='Question_ID', values='Value')
    
    # Plot the data
    pivot_table.plot(kind='bar', figsize=(10, 6))
    plt.title(f'Average Percentage of Women Agreeing with Each Question in {region}')
    plt.xlabel('Residence')
    plt.ylabel('Average Percentage')
    plt.legend(title='Question ID')
    plt.show()
```

### These graphs are more readble, but they can be confusing because in the y axis each has its own scale, it does not allow us to compare them. Let's solve it.

```python
num_regions = len(unique_regions)
# Create subplots
fig, axes = plt.subplots(num_regions, 1, figsize=(6, 4 * num_regions), sharey=True)
axes = axes.flatten()  # Flatten in case there's only one region

# Loop through each unique region to create separate plots
for i, region in enumerate(unique_regions):
    region_data = grouped[grouped['Region'] == region]
    pivot_table = region_data.pivot_table(index='Demographics Response', columns='Question_ID', values='Value')
    
    # Plot the data
    pivot_table.plot(kind='bar', ax=axes[i])
    axes[i].set_title(f'Average Percentage of Women Agreeing with Each Question in {region}')
    axes[i].set_xlabel('Residence')
    axes[i].set_ylabel('Average Percentage')
    axes[i].set_ylim(0, 100)  # Set y-axis limit to 100%
    axes[i].legend(title='Question ID')

plt.tight_layout()
plt.show()
```

## General Visualization <a class="anchor" id="General"></a>

```python
# Determine the number of rows and columns for the subplots
num_cols = 2
num_rows = (num_regions + num_cols - 1) // num_cols
# Create subplots
fig, axes = plt.subplots(num_rows, num_cols, figsize=(18, 6 * num_rows), sharey=True)
axes = axes.flatten()  # Flatten the axes array

# Define a color map for the question IDs based on the uploaded image
color_map = {
    1: '#1f77b4',  # Blue
    2: '#ff7f0e',  # Orange
    3: '#2ca02c',  # Green
    4: '#d62728',  # Red
    5: '#9467bd',  # Purple
    6: '#8c564b'   # Brown
}

# Loop through each unique region to create separate plots
for i, region in enumerate(unique_regions):
    region_data = grouped[grouped['Region'] == region]
    pivot_table = region_data.pivot_table(index='Demographics Response', columns='Question_ID', values='Value')
    
    # Plot the data
    ax = pivot_table.plot(kind='bar', ax=axes[i])
    axes[i].set_title(f'Average Percentage of Women Agreeing with Each Question in {region}')
    axes[i].set_xlabel('Residence')
    axes[i].set_ylabel('Average Percentage')
    axes[i].set_ylim(0, 100)  # Set y-axis limit to 100%
    axes[i].legend(title='Question ID')
    
    # Add annotations
    for container in axes[i].containers:
        axes[i].bar_label(container, fmt='%.0f%%', label_type='edge')

# Remove empty subplots
for j in range(i + 1, num_rows * num_cols):
    fig.delaxes(axes[j])

    
    axes[i].legend(title='Question ID')

# Add a legend for the question IDs and corresponding questions with colors
legend_labels = [f'{id}: {question}' for question, id in question_ids.items()]
handles = [plt.Line2D([0], [0], color=color_map[id], linewidth=0, marker='o', markersize=10) for id in question_ids.values()]
fig.legend(handles, legend_labels, loc='upper right', title='A husband is justified in hitting or beating his wife', bbox_to_anchor=(1.2, 1))

fig.savefig('average_percentage.png', bbox_inches='tight')
plt.tight_layout()
plt.show()
```

```python
# Loop through each unique region to create separate plots
for region in unique_regions:
    region_data = grouped[grouped['Region'] == region]
    pivot_table = region_data.pivot_table(index='Demographics Response', columns='Question_ID', values='Value')
    
    # Create a new figure for each region
    fig, ax = plt.subplots(figsize=(14, 6))
    
    # Plot the data
    ax = pivot_table.plot(kind='bar', ax=ax, color=[color_map[qid] for qid in pivot_table.columns])
    ax.set_title(f'Average Percentage of Women Agreeing with Each Question in {region}')
    ax.set_xlabel('Residence')
    ax.set_ylabel('Average Percentage')
    ax.set_ylim(0, 100)  # Set y-axis limit to 100%
    ax.legend(title='Question ID')
    
    # Add annotations
    for container in ax.containers:
        ax.bar_label(container, fmt='%.0f%%', label_type='edge')
    
 # Position the legend outside the plot
    ax.legend(handles, legend_labels, loc='center left', bbox_to_anchor=(1, 0.5), title='Question IDs')
    fig.savefig(f'{region}_average_percentage.png', bbox_inches='tight')
    plt.tight_layout(rect=[0, 0, 0.85, 1])  # Adjust layout to make space for the legend
    plt.show()
```

## Findings <a class="anchor" id="Findings2"></a>
These charts show very interesting insights. Here are som notable ones:
* Across all regions, the level of "agree" is observed to be higher in rural areas compared to urban areas.
* Notably, the regions with the lowest percentages of "agree" are South America (the lowest of all regions), the Balkans region, and other regions.
* Additionally, the regions with the highest percentages of "agree" are Africa, Asia, and the Middle East.

It reveals potential social and cultural differences between rural and urban areas, and how regions with distinct histories, cultures, religions and economies might hold varying attitudes.

# Conclusion <a class="anchor" id="Conclusion"></a>

The graphs illustrate that the proportion of women agreeing with the statement "a husband is justified in hitting or beating his wife" varies across regions and areas of residence. Generally, the proportion of "agree" is higher in rural areas compared to urban areas. The regions with the highest percentages of "agree" are Africa, Asia, and the Middle East, while the regions with the lowest percentages are South America, the Balkans, and Other regions.
